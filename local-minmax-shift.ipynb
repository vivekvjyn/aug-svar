```python
import numpy as np
import os
import matplotlib.pyplot as plt
from lib.utils import shift_change_points
```

```python
segment_dir = 'segment'
```

```python
i = 0
for segment_name in os.listdir(segment_dir):
    if i < 500:
        i += 1
        continue

    segment = np.load(os.path.join(segment_dir, segment_name))
    shifted = shift_change_points(segment, 10)

    for i in range(len(shifted)):
        plt.plot(shifted[i])
        
    break
```
